```python
#Bulan - pengeluaran permin

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
bulan = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Des']

Minggu1 = np.random.randint(200000,500001,12)
Minggu2 = np.random.randint(200000,500001,12)
Minggu3 = np.random.randint(200000,500001,12)
Minggu4 = np.random.randint(200000,500001,12)

data = {
    'Bulan_1' : bulan,
    'Minggu_1' : Minggu1,
    'Minggu_2' : Minggu2,
    'Minggu_3' : Minggu3,
    'Minggu_4' : Minggu4
}

df = pd.DataFrame(data)
df['Total_Pengeluaran'] = df[['Minggu_1','Minggu_2','Minggu_3','Minggu_4']].sum(axis=1)
df
```

```python
#Diagram Bar/Batang

plt.figure(figsize=(12,6)) #width,Height -> inch
plt.bar(df['Bulan'],df['Total_Pengeluaran'],color='red')
plt.title("Pengeluaran per Bulan")
plt.xticks(rotation=45)
plt.ylabel("Rupiah")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

for minggu in ['Minggu_1','Minggu_2','Minggu_3','Minggu_4']:
    plt.plot(df['Bulan'], df[minggu], marker='o', label=minggu)

plt.title('Pengeluaran per Minggu per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rupiah')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Diagram Pie Chart -> Proporsi / porsentase

plt.figure(figsize=(8,8))
plt.title('Proporsi Pengeluaran Per Bulan')
plt.pie(df['Total_Pengeluaran'],labels=df['Bulan'],startangle=90,autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"cpi.csv"
df = pd.read_csv(file_path, encoding='latin1')

df_2017 = df[['country', '2017']].dropna()
df_2017.columns = ['country', 'score_2017']
df_top10 = df_2017.sort_values(by='score_2017', ascending=False).head(10)
df_top5 = df_top10.head(5)

print("Top 10 Negara dengan Skor CPI Tertinggi pada 2017:")
print(df_top10.to_string(index=False))

plt.figure(figsize=(8, 6))
plt.pie(df_top5['score_2017'], labels=df_top5['country'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Negara dengan Skor CPI Tertinggi pada 2017')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
